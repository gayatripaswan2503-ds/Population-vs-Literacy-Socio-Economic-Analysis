```python
#Literacy Rate by State
plt.figure(figsize=(12,6))
sns.barplot(x='State', y='Literacy_Rate', data=data, palette='viridis')
plt.xticks(rotation=90)
plt.title("Literacy Rate by State in India")
plt.xlabel("State")
plt.ylabel("Literacy Rate (%)")
plt.show()
```

```python
#Youâ€™ll observe that states like  Kerala,Mizoram,Goa and Tripura
#show higher literacy rates, while  
#Jharkhand ,Bihar and  Andhra Pradesh have lower literacy rates.

plt.figure(figsize=(14,7))
sns.barplot(x='State', y='Literacy_Rate', hue='Year', data=data, palette=['#FF9933','#138808'])
plt.xticks(rotation=90)
plt.title('State-wise Literacy Rate Comparison (2011 vs 2021)')
plt.ylabel('Literacy Rate (%)')
plt.xlabel('State')
plt.show()
```

```python
#Literacy rates improved in every state between 2011â€“2021.
#Kerala maintains the highest literacy rate (95%).
#Bihar, Rajasthan, Uttar Pradesh show the strongest improvement (6â€“10% rise).
#Northern states lag slightly behind southern and western states in 2021.'''


#Per Capital income growth by state 
plt.figure(figsize=(14,7))
sns.barplot(y='State', x='Per_Capita_Income', hue='Year', data=data, palette='crest')
plt.title('Per Capita Income Growth by State (2011 vs 2021)')
plt.xlabel('Per Capita Income (â‚¹)')
plt.ylabel('State')
plt.show()
```

```python
#Goa,Sikkim and Telangana lead with the highest per capita income.
#Bihar, Assam, Uttar Pradesh remain lowest despite growth.
#Economic progress is strongest in Western and Southern India.



#Average Literacy of male and female
male = data[data['Year']==2011]['Male_Literacy'].mean()
female = data[data['Year']==2011]['Female_Literacy'].mean()

print("Average Male Literacy (2011):", round(male, 2))
print("Average Female Literacy (2011):", round(female, 2))
print("Average Literacy in 2021")
male = data[data['Year']==2021]['Male_Literacy'].mean()
female = data[data['Year']==2021]['Female_Literacy'].mean()

print("Average Male Literacy (2021):", round(male, 2))
print("Average Female Literacy (2021):", round(female, 2))
```

```python
# Filter 2021 data
df_2021 = data[data['Year'] == 2021]

# Region-wise averages
region_lit = df_2021.groupby('Region')[['Male_Literacy', 'Female_Literacy']].mean()

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12,6))

# Left Pie - Male Literacy
axes[0].pie(region_lit['Male_Literacy'], labels=region_lit.index, autopct='%1.1f%%',
            startangle=90, colors=plt.cm.Blues(range(50,250,30)))
axes[0].set_title('Region-wise Male Literacy (2021)')

# Right Pie - Female Literacy
axes[1].pie(region_lit['Female_Literacy'], labels=region_lit.index, autopct='%1.1f%%',
            startangle=90, colors=plt.cm.Purples(range(50,250,30)))
axes[1].set_title('Region-wise Female Literacy (2021)')

plt.suptitle('Comparison: Male vs Female Literacy by Region (2021)', fontsize=14)
plt.show()

```

```python
# West regions contribute the most to overall literacy for both genders.  
# North and East regions lag behind â€” female literacy particularly lower.




#Urban Population growth by states
plt.figure(figsize=(12,7))
sns.lineplot(x='State', y='Urban_Population', hue='Year', data=data, marker='o', palette='flare')
plt.xticks(rotation=90)
plt.title('Urban Population Growth by State (2011â€“2021)')
plt.ylabel('Urban Population (%)')
plt.xlabel('State')
plt.show()

```

```python
#Maharashtra and Karnatka are highly urbanized
#Rural-to-urban shift is strongest in Tamil Nadu, Gujarat, and Karnataka.
#Bihar and Assam remain highly rural, showing limited urban expansion.


#Employement Rate Comparision(2011 vs 2021)
plt.figure(figsize=(12,6))
sns.barplot(x='State', y='Employment_Rate', hue='Year', data=data, palette='viridis')
plt.xticks(rotation=90)
plt.title('Employment Rate Comparison (2011 vs 2021)')
plt.ylabel('Employment Rate (%)')
plt.xlabel('State')
plt.show()




```

```python
#Most states saw a 5â€“10% rise in employment.
#Kerala, Himachal Pradesh, and Sikkim show the highest employment levels.
#Some North-Eastern states remain below the national average despite literacy improvements.



#Education Vs Health Expenditure(2021)
plt.figure(figsize=(10,7))
sns.regplot(x='Education_Expenditure', y='Health_Expenditure',
            data=data[data['Year']==2021],
            scatter_kws={"color": "orange"}, line_kws={"color": "green"})

plt.title('Education vs Health Expenditure (2021)')
plt.xlabel('Education Expenditure (% of Total Budget)')
plt.ylabel('Health Expenditure (% of Total Budget)')

# Add state names to each dot
for i in range(len(data[data['Year']==2021])):
    row = data[data['Year']==2021].iloc[i]
    plt.text(row['Education_Expenditure']+0.05, row['Health_Expenditure']+0.05, 
             row['State'], fontsize=8)

plt.show()

```

```python
#A positive correlation exists â€” states investing more in education also invest more in health.
#Kerala and Goa stand out with high values for both.
#Bihar ,MP and Uttar Pradesh remain low on both metrics.'''

#Heat_Map
plt.figure(figsize=(10,6))
corr = data[data['Year']==2021][['Literacy_Rate','Employment_Rate','Per_Capita_Income','Urban_Population','Health_Expenditure','Education_Expenditure']].corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap (2021 Development Indicators)')
plt.show()



```

```python
#Literacy, Income, and Employment show strong positive correlation.
#Urbanization is linked to higher literacy and income.
#Education & Health Expenditure drive better socio-economic outcomes.'''
```
